```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import cv2
```

```python
%matplotlib inline
```

```python
GQA_Path = "/home/yifeng/Datasets/GQA/images/images/"
filenames = os.listdir(GQA_Path)
filenames.sort()
```

```python
Question_path = "/home/yifeng/Datasets/GQA/questions/train_balanced_questions.json"
f = open(Question_path,'r')
data = json.load(f)
```

```python
keys = list(data.keys())
```

```python
for key in keys:
    print(data[key])
    break
```

```python
# index = np.random.randint(0, len(filenames), 10)
s = 220
e = s+20
for key in keys[s:e]:
    path = GQA_Path + data[key]['imageId']+'.jpg'
    img = cv2.imread(path, 1)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.figure()
    plt.title(data[key]['imageId'])
    plt.text(0,-70, data[key]['question']+'/'+data[key]['answer'])
    plt.imshow(img)
```
